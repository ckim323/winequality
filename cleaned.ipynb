```python
import pandas as pd

#reads disorganized wine quality csv's and returns pandas dataframe 
redwine = pd.read_csv("winequality-red.csv")
whitewine = pd.read_csv("winequality-white.csv")
col_list = list(redwine.columns)

col_list[0] = col_list[0].replace('"', '')
col_list = col_list[0].split(';')

redwine.columns = ['temp']
whitewine.columns = ['temp']

redold = redwine.loc[:,'temp']
whiteold = whitewine.loc[:,'temp']
newred =[]
newwhite=[]
#newdf = pd.DataFrame(firstline,columns=col_list)
for row in redold:
    newred.append(row.split(';'))
for row in whiteold:
    newwhite.append(row.split(';'))

redwine_df = pd.DataFrame(newred,columns=col_list)
whitewine_df = pd.DataFrame(newwhite,columns=col_list)
```
